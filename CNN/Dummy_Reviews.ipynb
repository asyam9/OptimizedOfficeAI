```python
from faker import Faker
import random
import random
from datetime import date, timedelta
import pandas as pd

def random_date():
    start_date = date(2023, 6, 1)
    end_date = date(2023, 6, 30)
    
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date


def random_rating():
    return random.randint(1, 5)

fake = Faker('ko_KR')

title = [fake.catch_phrase() for i in range(500)]
data = [fake.catch_phrase() for i in range(500)]
ratings = [random_rating() for _ in range(500)]
dates = [random_date() for _ in range(500)]

Dummy_reviews = pd.DataFrame({'제목': title, '내용' : data, '별점' : ratings, '날짜' : dates})
Dummy_reviews.to_csv('Dummy_reviews.csv',index=False)
```
